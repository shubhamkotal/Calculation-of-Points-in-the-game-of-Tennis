```python
data1=data1.drop('outcome',axis=1)
```

```python
#dummification
data3=pd.get_dummies(data1)
data3.dtypes
```

```python
#split data
from sklearn import preprocessing, metrics, cross_validation
from sklearn.model_selection import train_test_split
```

```python
# Divide in to train and test
train_target=data["outcome"]
#train=data.drop('outcome', axis=1)
#from sklearn.model_selection import train_test_split  
train1,test1,train_target1,test_target1 = train_test_split(data3, train_target, test_size=0.20,shuffle=False)  
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
train1.dtypes
```

```python
data2=data3.select_dtypes(['int64','float64']).columns
# standardizing
scaler = StandardScaler()
scaler.fit(train1[data2])
```

```python
#transforming
train1[data2]=scaler.transform(train1[data2])
test1[data2]=scaler.transform(test1[data2])
```

```python
train1.shape
```

```python
#model1 on adaboost
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
Adaboost_model = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=6),
    n_estimators=1000,
    learning_rate=0.001)


Adaboost_model.fit(train1, train_target1)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score

train_pred1 = Adaboost_model.predict(train1)
print(accuracy_score(train_target1,train_pred1))
```

```python
test_pred1= Adaboost_model.predict(test1)
print(accuracy_score(test_target1,test_pred1))
```
